# Static train ridership on the Swedish railways 2024.

This is a static analysis of train ridership on the Swedish railways during 2024!

## Read Sampers data

```python
import pandas as pd # type: ignore

# Read the CSV file into a DataFrame
df_pax = pd.read_csv('../data/cleaned_data/Sampers-pax-2024-cleaned.csv', sep=';', low_memory=False, dtype={"line": str})  # Force 'line' column as string

# Define a function to replace commas with dots and remove hidden characters
def clean_column_values(value):
    if isinstance(value, str):
        value = value.replace(',', '.')  # Replace commas with dots for decimals
        value = value.replace('\r', '').replace('\n', '')  # Remove carriage returns and line breaks
    return value

# Apply cleaning to all columns
for column in df_pax.columns:
    if df_pax[column].dtype == 'object':
        df_pax[column] = df_pax[column].apply(clean_column_values)

# Convert all columns except the first three to numeric
numeric_columns = df_pax.columns[2:]  # Select all columns starting from the 3th (index 2)
df_pax[numeric_columns] = df_pax[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

## Passengers ombord

A new data structure where the number of passengers ombord is calculated.

```python
# Columns for boardings and alightings
boarding_cols = ["Nat pr PÅ", "Nat tj PÅ", "Nat arb PÅ", "Reg arb PÅ", "Reg tj PÅ", "Reg övr PÅ"]
alighting_cols = ["Nat pr AV", "Nat tj AV", "Nat arb AV", "Reg arb AV", "Reg tj AV", "Reg övr AV"]

# Ensure the relevant columns are floats
df_pax[boarding_cols] = df_pax[boarding_cols].astype(float)
df_pax[alighting_cols] = df_pax[alighting_cols].astype(float)

# Shift the values of alighting columns by one row within each group and replace missing values with zeros
df_pax[alighting_cols] = df_pax.groupby('line')[alighting_cols].shift(+1).fillna(0)
```

```python
# for each line, compare the total number of boardings and alightings
# if the total number of boardings is greater than the total number of alightings, print the line
for line in df_pax['line'].unique():
    df_line = df_pax[df_pax['line'] == line]
    total_boardings = df_line[boarding_cols].sum().sum()
    total_alightings = df_line[alighting_cols].sum().sum()
    if abs(total_boardings - total_alightings)>= 1e-3:
        print(f"Line {line} has more boardings ({total_boardings}) than alightings ({total_alightings})")
```

```python
# Create masks for non-zero boardings and alightings
has_boardings = df_pax[boarding_cols].sum(axis=1) > 0
has_alightings = df_pax[alighting_cols].sum(axis=1) > 0

# Function to keep the first and last row of each group and specific rows based on conditions
def keep_first_last(group):
    mask = has_boardings.loc[group.index] | has_alightings.loc[group.index]
    mask.iloc[0] = True  # Ensure the first row is kept
    mask.iloc[-1] = True  # Ensure the last row is kept

    # Ensure specific rows are kept based on conditions
    conditions = [
        ((group['line'] == '3001') | (group['line'] == '3001R')) & (group['Station'] == 'Björkliden'),
        ((group['line'] == '40010') | (group['line'] == '40010R')) & (group['Station'] == 'Björkliden'),
        ((group['line'] == '4101') | (group['line'] == '4101R')) & (group['Station'] == 'Umeå Östra'),
        ((group['line'] == '4203') | (group['line'] == '4203R')) & (group['Station'] == 'Åre'),
        ((group['line'] == '4203') | (group['line'] == '4203R')) & (group['Station'] == 'Duved'),
        ((group['line'] == '5401') | (group['line'] == '5401R')) & (group['Station'] == 'Torsåker'),
        ((group['line'] == '7501') | (group['line'] == '7501R')) & (group['Station'] == 'Grythyttan')
    ]
    for condition in conditions:
        mask |= condition

    return group[mask]

# Apply the function to each group
df_pax = df_pax.groupby('line', group_keys=False).apply(keep_first_last).reset_index(drop=True)
```

```python
import re

# Remove duplicate columns if any
df_pax = df_pax.loc[:, ~df_pax.columns.duplicated()]

# Handle unmatched stations using regex
unmatched_stations = df_pax[df_pax['Station_sign'].isna()]['Station'].unique()

# define a function to match stations based on regex
def match_station(station, plats_list):

    # Prepare station names by cleaning and applying variants
    station_variants = [
        station,  # original name
        re.sub(r'central', 'c', station, flags=re.IGNORECASE),  # Replace "central" with "c"
        re.sub(r'(\w+)( central)', r'\1s central', station, flags=re.IGNORECASE),  # Add "s" before "central"
        re.sub(r'(\w+)( c)', r'\1s c', station, flags=re.IGNORECASE),  # Add "s" before "c"
        # remove 's' at the end of the station name before "central"
        re.sub(r'(\w+)s central', r'\1 central', station, flags=re.IGNORECASE),
        # remove 's' at the end of the station name before "central" and replace "central" with "c"
        re.sub(r'(\w+)s central', r'\1 c', station, flags=re.IGNORECASE),
    ]

    # Try to match each variant against plats_list
    for variant in station_variants:
        for plats in plats_list:
            if re.match(re.escape(variant), plats, re.IGNORECASE):
                return plats
    return None
```

```python
# Apply the function to unmatched stations
for station in unmatched_stations:
    match = match_station(station, df_plats_sign['Plats'])
    if match:
        df_pax.loc[df_pax['Station'] == station, 'Station_sign'] = df_plats_sign.loc[df_plats_sign['Plats'] == match, 'Plats_sign'].values[0]
        # Add the matched station to df_plats_sign
        new_row = pd.DataFrame({'Plats': [station], 'Plats_sign': [df_plats_sign.loc[df_plats_sign['Plats'] == match, 'Plats_sign'].values[0]]})
        df_plats_sign = pd.concat([df_plats_sign, new_row], ignore_index=True)
```

Now we can create our dataframe using signatures instead of station names.

```python
# Initialize a list to store the results
data = []

# Group by 'line' and process each group separately
for line, group in df_pax.groupby("line"):
    # Reset the index for each group to ensure sequential processing
    group = group.reset_index(drop=True)

    # Initialize onboard counters for each category of passengers
    onboard = {col: 0 for col in boarding_cols}

    for i in range(len(group) - 1):
        # Add boardings at the current station
        for col in boarding_cols:
            onboard[col] += group.loc[i, col]

        # Round and ensure counts are >= 0
        #rounded_onboard = {k: max(round(v), 0) for k, v in onboard.items()}
        rounded_onboard = {k: max(v, 0) if v > 1e-6 else 0 for k, v in onboard.items()}
        # Without rounding nor ensuring counts are >= 0
        #rounded_onboard = onboard

        # Append the data for the current segment (from the current station to the next)
        data.append({
            "Linje": line,
            # the first and last station of the line
            "först_sign": group.loc[0, "Station_sign"] if pd.notna(group.loc[0, "Station_sign"]) else group.loc[0, "Station"],
            "sist_sign": group.iloc[-1]["Station_sign"] if pd.notna(group.iloc[-1]["Station_sign"]) else group.iloc[-1]["Station"],
            # the first and last station of the line (original names)
            "först": group.loc[0, "Station"],
            "sist": group.iloc[-1]["Station"],
            # the pair of stations for the current segment
            "från_sign": group.loc[i, "Station_sign"] if pd.notna(group.loc[i, "Station_sign"]) else group.loc[i, "Station"],
            "till_sign": group.loc[i+1, "Station_sign"] if pd.notna(group.loc[i+1, "Station_sign"]) else group.loc[i+1, "Station"],
            # the pair of stations for the current segment (original names)
            "från": group.loc[i, "Station"],
            "till": group.loc[i+1, "Station"],
            # the number of passengers onboard for the current segment
            "ombord_pr_nat": rounded_onboard["Nat pr PÅ"],
            "ombord_tj_nat": rounded_onboard["Nat tj PÅ"],
            "ombord_arb_nat": rounded_onboard["Nat arb PÅ"],
            "ombord_arb_reg": rounded_onboard["Reg arb PÅ"],
            "ombord_tj_reg": rounded_onboard["Reg tj PÅ"],
            "ombord_övr_reg": rounded_onboard["Reg övr PÅ"],
            # the number of passengers alighting at the next station
            "avstigande_pr_nat": group.loc[i+1, "Nat pr AV"],
            "avstigande_tj_nat": group.loc[i+1, "Nat tj AV"],
            "avstigande_arb_nat": group.loc[i+1, "Nat arb AV"],
            "avstigande_arb_reg": group.loc[i+1, "Reg arb AV"],
            "avstigande_tj_reg": group.loc[i+1, "Reg tj AV"],
            "avstigande_övr_reg": group.loc[i+1, "Reg övr AV"],
            # the number of passengers boarding at the current station
            "påstigande_pr_nat": group.loc[i, "Nat pr PÅ"],
            "påstigande_tj_nat": group.loc[i, "Nat tj PÅ"],
            "påstigande_arb_nat": group.loc[i, "Nat arb PÅ"],
            "påstigande_arb_reg": group.loc[i, "Reg arb PÅ"],
            "påstigande_tj_reg": group.loc[i, "Reg tj PÅ"],
            "påstigande_övr_reg": group.loc[i, "Reg övr PÅ"]
        })

        # Subtract alightings for the current station (these apply to the next segment)
        for b_col, a_col in zip(boarding_cols, alighting_cols):
            onboard[b_col] -= group.loc[i+1, a_col]

# Convert the list to a DataFrame
new_df_pax = pd.DataFrame(data)
# Filter out rows where the station name is "0" or "A6"
#new_df_pax = new_df_pax[(new_df_pax["från"] != "0") | (new_df_pax["från"] != "A6")].reset_index(drop=True)
```

These are total numbers per day, we need to divide per number of round trips. We use data from Linjer in excel file Sampers_input_data_2024.xlsx.

```python
### read sheet Linjer from the excel file Sampers_input_data_2024.xlsx
df_linjer = pd.read_excel('../data/cleaned_data/Sampers_input_data_2024_min_plats.xlsx', sheet_name='Linjer', header=0)

# Column Antal dubbel-turer (ut2) gives the number of round trips for each Linje.
# Divide the number of passengers boarding, alighting and onboard by the number of round trips.
# This will give the number of passengers per round trip.
df_linjer['Antal dubbel-turer (ut2)'] = df_linjer['Antal dubbel-turer (ut2)'].astype(int)
#df_linjer['Antal dubbel-turer (ut2)'] = df_linjer['Antal dubbel-turer (ut2)'].replace(0, 1) # Replace 0 with 1 to avoid division by zero

# make sure column Linje is of type string
df_linjer['Linje'] = df_linjer['Linje'].astype(str)

# create a list of columns containing the number of passengers
passenger_cols = ["ombord_pr_nat", "ombord_tj_nat", "ombord_arb_nat", "ombord_arb_reg", "ombord_tj_reg", "ombord_övr_reg",
                  "avstigande_pr_nat", "avstigande_tj_nat", "avstigande_arb_nat", "avstigande_arb_reg", "avstigande_tj_reg", "avstigande_övr_reg",
                  "påstigande_pr_nat", "påstigande_tj_nat", "påstigande_arb_nat", "påstigande_arb_reg", "påstigande_tj_reg", "påstigande_övr_reg"]

# For each Linje, update the columns of number of passengers boarding, alighting and onboard in new_df_pax, i.e., 6th column to last
# by dividing the values by the number of round trips.
for i in range(len(df_linjer)):
    linje = df_linjer.loc[i, 'Linje']
    antal_dubbel_turer = df_linjer.loc[i, 'Antal dubbel-turer (ut2)']
    new_df_pax.loc[new_df_pax['Linje'] == linje, passenger_cols] = new_df_pax.loc[new_df_pax['Linje'] == linje, passenger_cols] / antal_dubbel_turer
```

```python
## for each Linje add a column with the average speed.
## Use the running time in minutes from column Linjetid_minuter and the distance in km from column  Linjelängd_kilometer
## to calculate the average speed in km/h
df_linjer['Linjetid_minuter'] = df_linjer['Linjetid_minuter'].astype(float)
df_linjer['Linjelängd_kilometer'] = df_linjer['Linjelängd_kilometer'].astype(float)
df_linjer['Medelhastighet'] = df_linjer['Linjelängd_kilometer'] / (df_linjer['Linjetid_minuter'] / 60)
```

```python
# do the same but plot the line and its inverse together
# Get a random line

import matplotlib.pyplot as plt
import numpy as np

# Get a random line that has an inverse (remove 'R' suffixed lines from selection)
base_lines = [line for line in new_df_pax['Linje'].unique() if not line.endswith('R')]
random_line = np.random.choice(base_lines)
inverse_line = random_line + 'R'

# Get data for both lines
line_data = new_df_pax[new_df_pax['Linje'] == random_line]
inverse_data = new_df_pax[new_df_pax['Linje'] == inverse_line]

# Calculate total passengers onboard
onboard_cols = ['ombord_pr_nat', 'ombord_tj_nat', 'ombord_arb_nat', 
                'ombord_arb_reg', 'ombord_tj_reg', 'ombord_övr_reg']
total_onboard = line_data[onboard_cols].sum(axis=1)
total_onboard_inverse = inverse_data[onboard_cols].sum(axis=1).iloc[::-1].reset_index(drop=True)
# total_onboard_inverse should be mirroed because the data is for the inverse line

# Create segment labels
segments = [f"{row['från']} - {row['till']}" for _, row in line_data.iterrows()]
segments_inverse = [f"{row['från']} - {row['till']}" for _, row in inverse_data.iterrows()]

plt.figure(figsize=(12, 6))
plt.plot(range(len(segments)), total_onboard, marker='o', label=f"{random_line} ({line_data['först'].unique()[0]} - {line_data['sist'].unique()[0]})")
plt.plot(range(len(segments_inverse)), total_onboard_inverse, marker='s', label=f"{inverse_line} ({inverse_data['först'].unique()[0]} - {inverse_data['sist'].unique()[0]})")
plt.xticks(range(max(len(segments), len(segments_inverse))), segments, rotation=45, ha='right')
plt.ylabel('Total Passengers Onboard')
plt.title(f'Passenger Variation: {random_line} vs {inverse_line}')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
